```python
import pandas as pd

# Replace 'rt.reviews.tsv' with the name of your TSV file
df = pd.read_csv('rt.reviews.tsv', delimiter='\t', encoding='ISO-8859-1')

# Display the contents of the file
df.head(10)

```

```python
#drop any colums that are not necessary
df = df.drop(['critic', 'top_critic', 'publisher'], axis=1)
df.head()
```

```python
#convert the date column to date time data type
df['date'] = pd.to_datetime(df['date'], format='%B %d, %Y')
df.head()
```

```python
#clean the rating column
df['rating'] = pd.to_numeric(df['rating'].str.replace(r'[^\d.]+', ''), errors='coerce')
```

```python
df.head()
```

```python
#drop any missing raws
df = df.dropna()
df.head()
```

```python
# Convert the 'review' column to a numerical data type
df['review'] = pd.to_numeric(df['review'], errors='coerce')

# Calculate the correlation between the 'rating' and 'review' columns
corr = df['rating'].corr(df['review'], min_periods=1)

# Print the correlation coefficient
print("Correlation coefficient between 'rating' and 'review':", corr)
```

After testing the correlation between reviews and rating there was no correlation thats why we got the "NAN' output. where 'NAN' occurs if the correlation coefficient between the 'rating' and 'review' columns is NaN, it may indicate that there is no linear relationship between these variables in the dataset. It could also mean that the 'review' column contains missing or invalid values, which are being converted to NaN by the pd.to_numeric() function. In this case, we had dropped the missing values hence making sure the error doesnt occur.
